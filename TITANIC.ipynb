```python
X=datos.drop(["Survived"], axis=1) #separa el 80% y el 20%
y=datos["Survived"]
from sklearn.model_selection import train_test_split
X_ent,X_pru, y_ent, y_pru=train_test_split(X,y, test_size=.2)
```

```python
from sklearn.linear_model import LogisticRegression
modelo=LogisticRegression() #cuando el modelo marca el numero de iteraciones 
modelo.fit(X_ent,y_ent)
```

```python
predicciones=modelo.predict(X_pru) #calculamos las predicciones
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pru, predicciones)  #se checa que tan bueno es el modelo de prediccion eficiencia
```
